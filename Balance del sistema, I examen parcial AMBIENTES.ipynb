### Balance del sistema

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
#CONDENSADOR
cond_cap=[67532,77402,88215,100062,113038,127234,142742,159657,178070,198073,219761,243244]
cond_temp_succ=np.arange(-10,50,5)
```

```python
#EVAPORADOR
#serie del diferencial de temperatura se otiene como la diferencia entre la temperatura de succión y 
#el diferencial seleccionado que es de 16 °F
DT_series=32-cond_temp_succ;
#lA capacidad se obtiene al multiplicar el diferencial por la tasa de cambio de la capacidad en función del DT. Es decir
# Cap= cte * DT, donde cte es la capacidad de la tabla del fabricante dividida entreel DT respectivo según especificación
#en este caso cte=190000[Btu/h]/20°F
evap_cap=(95000/10)*DT_series
```

```python
a,ax=plt.subplots(num=1,
    figsize=(12,12),
    facecolor="darkorange",
    edgecolor="black")
plt.plot(cond_temp_succ,cond_cap)
plt.plot(cond_temp_succ,evap_cap)
ax.set_xlabel("Temperatura de saturación en la succión [20°F]",fontsize=14)
ax.set_ylabel("Capacidad [Btu/h]",fontsize=14)
ax.grid()
y_etiq=np.arange(min(evap_cap),max(evap_cap),20000)
ax.set_yticks(y_etiq)
x_etiq=np.arange(min(cond_temp_succ),max(cond_temp_succ),2)
ax.set_xticks(x_etiq)
```
